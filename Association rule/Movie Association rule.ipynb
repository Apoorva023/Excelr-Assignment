```python
#Import libraries

import pandas as pd
from mlxtend.frequent_patterns import association_rules,apriori
from mlxtend.preprocessing import TransactionEncoder
import warnings
warnings.filterwarnings('ignore')
```

```python
#Load data
df = pd.read_csv('my_movies.csv')
df
```

```python
data = df.iloc[:,5:]
```

```python
# Apply the Apriori algorithm to find frequent itemsets with min_support=0.1
score1 = apriori(data, min_support=0.1,use_colnames=True)
score1
```

```python
# Apply the Apriori algorithm to find frequent itemsets with min_support=0.2
score2 = apriori(data, min_support=0.2,use_colnames=True)
score2
```

```python
# Apply the Apriori algorithm to find frequent itemsets with min_support=0.3
score3 = apriori(data, min_support=0.3,use_colnames=True)
score3
```

```python
# Apply the Apriori algorithm to find frequent itemsets with min_support=0.4
score4 = apriori(data, min_support=0.4,use_colnames=True)
score4
```

```python
#Model Building

model= association_rules(score2)
model
```

```python
#print only lift value more than 1, to find the good model
model[model['lift']>1]
```

```python
Visualization
```

```python
import matplotlib.pyplot as plt

# Plotting support vs. confidence scatter plot
plt.scatter(model['support'], model['confidence'], s=model['lift']*100, alpha=0.5)
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.title('Support vs. Confidence Scatter Plot')
plt.show()
```

```python
import networkx as nx

# Creating a directed graph
G = nx.DiGraph()

# Adding nodes and edges to the graph
for idx, row in model.iterrows():
    G.add_edge(', '.join(row['antecedents']), ', '.join(row['consequents']), weight=row['confidence'])

# Plotting the network graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G)
nx.draw_networkx(G, pos, with_labels=True, node_size=5000, edge_color='gray', alpha=0.8, font_size=10, width=[d['weight'] * 5 for (u, v, d) in G.edges(data=True)])
plt.title('Association Rules Network Graph')
plt.axis('off')
plt.show()
```
